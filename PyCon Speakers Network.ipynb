```python
import csv
import networkx as nx
import matplotlib.pylab as plt

# list of PyCon17 speakers
pycon17 = open("pycon17speakers.txt","r")
reader = csv.reader(pycon17)
pycon17list = [row for row in reader]
py17spkrs = [ s[0] for s in pycon17list]

# list of PyCon18 speakers
pycon18 = open("pycon18speakers.txt","r")
reader = csv.reader(pycon18)
pycon18list = [row for row in reader]
py18spkrs = [ s[0] for s in pycon18list]


# graph for PyCon 17
g17 = nx.Graph()
g17 = nx.complete_graph(len(py17spkrs), nx.Graph())
py17labels = dict(zip(range(1,len(py17spkrs)),py17spkrs))
g17 = nx.relabel_nodes(g17, py17labels)
nx.draw(g17, show_labels=True)
plt.show()

# graph for PyCon18
g18 = nx.Graph()
g18 = nx.complete_graph(len(py18spkrs), nx.Graph())
py18labels = dict(zip(range(1,len(py18spkrs)),py18spkrs))
g18 = nx.relabel_nodes(g18, py18labels)
nx.draw(g18, show_labels=True)
plt.show()

combG = nx.compose(g18,g17)
##pos = nx.kamada_kawai_layout(combG)
pos = nx.circular_layout(combG)
labels = {}
for idx, node in enumerate(combG.node()):
	labels[node] = node


nx.draw(combG,pos, node_color="blue", edge_color="yellow")
nx.draw_networkx_labels(combG,pos,labels, font_size=5)
plt.show()

```
